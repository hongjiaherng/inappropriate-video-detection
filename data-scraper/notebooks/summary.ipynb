```python
import matplotlib.pyplot as plt
%matplotlib inline

import youtube
import tiktok
```

## Verify Download

```python
yt_scraper = youtube.YoutubeScraper(root_path=root_path, n_videos_per_keyword_per_lang=3)
df_by_keyword_yt, df_by_lang_yt = yt_scraper.verify_download()
df_by_keyword_yt = df_by_keyword_yt.drop(labels=["TOTAL"])
df_by_lang_yt = df_by_lang_yt.drop(labels=["TOTAL"])
```

```python
tt_scraper = tiktok.TikTokScraper(root_path=root_path, n_videos_per_keyword_per_lang=3)
df_by_keyword_tt, df_by_lang_tt = tt_scraper.verify_download()
df_by_keyword_tt = df_by_keyword_tt.drop(labels=["TOTAL"])
df_by_lang_tt = df_by_lang_tt.drop(labels=["TOTAL"])
```

## Summary

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.bar(list(df_by_keyword_yt.keys()), list(df_by_keyword_yt.values))
ax1.set_xticks(range(len(df_by_keyword_yt)))
ax1.set_xticklabels(list(df_by_keyword_yt.keys()), rotation=90)  # Rotate x-axis labels for better visibility
ax1.set_xlabel('Keyword')
ax1.set_ylabel('Count')
ax1.set_title(f'Videos collected from YouTube Shorts (Total={sum(df_by_keyword_yt.values)})')

ax2.bar(list(df_by_keyword_tt.keys()), list(df_by_keyword_tt.values))
ax2.set_xticks(range(len(df_by_keyword_tt))) 
ax2.set_xticklabels(list(df_by_keyword_tt.keys()), rotation=90)  # Rotate x-axis labels for better visibility
ax2.set_xlabel('Keyword')
# ax2.set_ylabel('Count')
ax2.set_title(f'Videos collected from TikTok (Total={sum(df_by_keyword_tt.values)})')

# Adjust the layout to prevent labels from being cut off
plt.tight_layout()

# Display the combined plot
plt.show()
```

```python
lang_map = {
    "en": "English",
    "es": "Spanish",
    "pt": "Portuguese",
    "hi": "Hindi",
    "ko": "Korean",
    "zh-cn": "Simplified Chinese",
    "ms": "Malay",
}

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 4))

ax1.bar(list(df_by_lang_yt.keys()), list(df_by_lang_yt.values))
ax1.set_xticks(range(len(df_by_lang_yt)))
ax1.set_xticklabels([lang_map[lang] for lang in df_by_lang_yt.keys()], rotation=60)  # Rotate x-axis labels for better visibility
ax1.set_xlabel('Language')
ax1.set_ylabel('Count')
ax1.set_title(f'Videos collected from YouTube Shorts (Total={sum(df_by_lang_yt.values)})')

ax2.bar(list(df_by_lang_tt.keys()), list(df_by_lang_tt.values))
ax2.set_xticks(range(len(df_by_lang_tt))) 
ax2.set_xticklabels([lang_map[lang] for lang in df_by_lang_tt.keys()], rotation=60)  # Rotate x-axis labels for better visibility
ax2.set_xlabel('Language')
# ax2.set_ylabel('Count')
ax2.set_title(f'Videos collected from TikTok (Total={sum(df_by_lang_tt.values)})')

# Adjust the layout to prevent labels from being cut off
plt.tight_layout()

# Display the combined plot
plt.show()
```
